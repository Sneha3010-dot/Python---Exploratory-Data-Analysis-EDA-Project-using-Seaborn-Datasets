```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datasets = sns.get_dataset_names()
print(datasets)
```

```python
df = sns.load_dataset('car_crashes')
print(df.head())
```

```python
#### Which states have the highest and lowest accident rates?
sns.barplot(x = 'abbrev', y ='total', data=df)
```

```python
#### Is there a relationship between speeding and total crashes?
sns.scatterplot(x = 'speeding', y='total',data = df)
```

```python
#### Does insurance coverage percentage affect fatal crash rates?
sns.scatterplot(x = 'ins_premium', y='total',data = df)
```

```python
#### How do alcohol-related crashes correlate with total crashes?
corr = df[['total','alcohol']].corr()
print(corr)
```

```python
sns.heatmap(corr, annot = True)
```

```python
#### Identify outliers and discuss possible reasons.
sns.boxplot(data=df[['total','speeding','alcohol','ins_premium']])
```
