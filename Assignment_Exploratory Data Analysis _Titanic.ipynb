```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
datasets = sns.get_dataset_names()
print(datasets)
```

```python
df = sns.load_dataset('titanic')
print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
#### What percentage of passengers survived the disaster?
survived_count = df['survived'].value_counts()[1]
print(f"Number of passengers who survived: {survived_count}")

survival_rate = df['survived'].mean() * 100

print(f"Percentage of passengers who survived: {survival_rate:.2f}%")

sns.countplot(x='survived', data= df)
```

```python
#### How does survival rate vary by gender and passenger class?
sns.barplot(x='class', y='survived', hue='sex', data= df)
```

```python
#### Are younger passengers more likely to survive?
sns.boxplot(x='survived', y='age', data= df)
```
